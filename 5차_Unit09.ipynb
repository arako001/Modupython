Unit09. 파이차트 그리기

```python
import csv
f=open('gender.csv')
data=csv.reader(f)

m=[]
f=[]

city=input('원하는 지역: ')

for row in data:
    if city in row[0]:
        for i in range(0,101):
            m.append(row[i+3])
            f.append(row[-(i+1)])
f.reverse()
            
m=[-int(i.replace(",","")) for i in m]
f=[int(i.replace(",","")) for i in f]


import matplotlib.pyplot as plt
plt.rcParams['axes.unicode_minus']='False'
plt.rc('font',family='Malgun Gothic')
plt.title(city+'성별 분포')
plt.barh(range(len(m)),m, label='male')
plt.barh(range(len(f)),f, label='female')
```

```python
import csv
f=open('gender.csv')
data=csv.reader(f)

m=[]
f=[]

region=input('동 정보: ')
city=input('도시 정보: ')

for row in data:
    if region in row[0]:
        if city in row[0]:
            for i in range(0,101):
                m.append(row[i+3])
                f.append(row[-(i+1)])
f.reverse()

m=[-int(i.replace(",","")) for i in m]
f=[int(i.replace(",","")) for i in f]

plt.barh(range(len(m)),m,label='male')
plt.barh(range(len(f)),f,label='female')
plt.legend()
plt.show()
```
