Unit10. 산점도

```python
#성별 연령별 분포 비교
import csv
f=open('gender.csv')
data=csv.reader(f)

m=[]
f=[]

city=input('원하는 지역: ')

for row in data:
    if city in row[0]:
        for i in range(0,101):
            m.append(row[i+3]) #남성은 4번째(5열)부터. replace로 comma없앤 후, int로 변환 필요. 이 라인에서는 안되더라..
            f.append(row[-(i+1)]) #여성은 -1번째부터
        break
f.reverse()

m=[int(i.replace(",","")) for i in m] # m=int([i.replace(",","")for i in m]) 실행 시 에러. int는 list에 먹히지 않음. 
f=[int(i.replace(",","")) for i in f]

import matplotlib.pyplot as plt
plt.title('2020년 3월 '+city+' 성별 인구분포')
plt.rc('font',family='Malgun Gothic') #안쓰면 에러남
plt.rcParams['axes.unicode_minus']=False #음수 읽기
plt.plot(m,label='Male',color='yellow')
plt.plot(f,label='Female',color='purple')
plt.legend()
plt.show()
```

막대그래프 그리기

```python
import csv
f=open('gender.csv')
data=csv.reader(f)

result=[]

city=input('원하는 지역: ')

for row in data:
    if city in row[0]:
        for i in range(3,104):
            result.append(int(row[i].replace(",",""))- int(row[i+101].replace(",","")))
        break

print(result)

import matplotlib.pyplot as plt
plt.title('2020년 3월 '+city+'남성과 여성 인구 차이 분포')
plt.rc('font',family='Malgun Gothic') #안쓰면 에러남
# plt.rcParams['axes.unicode_minus']=False #음수 읽기
plt.bar(range(101),result)
plt.show()
```

연령대별 성별 비율 산점도로 그리기

```python
import csv
f=open('gender.csv')
data=csv.reader(f)

m=[]
f=[]

city=input('원하는 지역: ')

for row in data:
    if city in row[0]:
        for i in range(3,104):
            m.append(int(row[i].replace(",","")))
            f.append(int(row[i+101].replace(",","")))
        break

import matplotlib.pyplot as plt
plt.rc('font',family='Malgun Gothic') #안쓰면 에러남
plt.title('2020년 3월 '+city+'인구 분포')
# plt.rcParams['axes.unicode_minus']=False #음수 읽기
plt.scatter(m,f,c=range(101),alpha=0.5)
plt.colorbar()
plt.plot(range(max(m)),range(max(f)),'g') #추세선 추가
plt.show()
```
