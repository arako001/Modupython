```python
import csv
import matplotlib.pyplot as plt
f=open('seoul.csv')
data=csv.reader(f)
next(data)
aug=[]
jan=[]


for row in data:
    month=row[0].split('-')[1]
    if row[-1] !='':
        if month=='08':
            aug.append(float(row[-1]))
        if month=='01':
            jan.append(float(row[-1]))

plt.hist(sorted(aug),bins=100,color='r',label='Aug')
plt.hist(sorted(jan),bins=100,color='b',label='Jan')
plt.legend()
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt
f=open('seoul.csv')
data=csv.reader(f)
next(data)
month=[]
for i in range(12):
    month.append([])

    
for row in data:
    if row[-1] !='':
        month[int(row[0].split('-')[1])-1].append(float(row[-1]))

plt.style.use('ggplot')        
plt.figure(figsize=(10,5), dpi=300)            
plt.boxplot(month)
plt.show()
```
